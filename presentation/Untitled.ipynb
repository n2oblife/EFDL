```python
import matplotlib.pyplot as plt
import numpy as np

X = np.array([13.4,13.4,13.4, 20, 20,20])
Y = np.array([69.9,77.8,79,10.2,9.8,9.9])

plt.scatter(X,Y)
plt.xlabel("Number of parameters (M)")
plt.ylabel("Accuracy(%)")
plt.show()

plt.plot(13.4, 69.9, label="VGG-16, 10 epochs")
plt.plot(13.4, 77.8, label="VGG-16, 30 epochs")
plt.plot(13.4, 79, label="VGG-16, 50 epochs")
plt.plot(20, 10.2, label="VGG-19, batch size of 32")
plt.plot(20, 9.8, label="VGG-19, batch size of 64")
plt.plot(20, 9.9, label="VGG-19, batch size of 512")
plt.show()
```
